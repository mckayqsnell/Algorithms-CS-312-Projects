```python
import matplotlib.pyplot as plt

# Assuming you have the data stored in two lists: n_values and t_values
# Replace n_values and t_values with your actual data

# Data
n_values = [10, 100, 1000, 10000, 100000, 500000, 1000000]
t_values = [0.000, 0.005, 0.0624, 0.5004, 4.8546, 27.0548, 54.7226]

plt.plot(n_values, t_values)
plt.xscale('log')  # Set logarithmic scale for x-axis (n)
plt.xlabel('n')
plt.ylabel('t')
plt.title('Plot of n versus t')
plt.grid(True)
plt.show()
```
